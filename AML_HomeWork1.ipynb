# Question 2 Part 2:

```python
import matplotlib.pyplot as plt 
plt.style.use('dark_background') 
import numpy as np 
noise_scale = 100  
number_of_samples = 50
x =  25*(np.random.rand(number_of_samples, 1) - 0.8) 
y = 5 * x + 20 * x**2 + 1 * x**3 + noise_scale*np.random.randn(number_of_samples, 1) 
flatx = x.flatten()
flaty = y.flatten()
m1=np.polyfit(flatx,flaty,1) #1
MeanSq1 = np.square(np.subtract(np.polyval(m1,x),y).mean())
print("Mean Square Error 1:",MeanSq1)
m2=np.polyfit(flatx,flaty,2) #2
MeanSq2 = np.square(np.subtract(np.polyval(m2,x),y).mean())
print("Mean Square Error 2:",MeanSq2)
m3=np.polyfit(flatx,flaty,3) #3
MeanSq3 = np.square(np.subtract(np.polyval(m3,x),y).mean())
print("Mean Square Error 3:",MeanSq3)
m4=np.polyfit(flatx,flaty,4) #4
MeanSq4 = np.square(np.subtract(np.polyval(m4,x),y).mean())
print("Mean Square Error 4:",MeanSq4)
m5=np.polyfit(flatx,flaty,5) #5
MeanSq5 = np.square(np.subtract(np.polyval(m5,x),y).mean())
print("Mean Square Error 5:",MeanSq5)
m6=np.polyfit(flatx,flaty,6) #6
MeanSq6 = np.square(np.subtract(np.polyval(m6,x),y).mean())
print("Mean Square Error 6:",MeanSq6)
m7=np.polyfit(flatx,flaty,7) #7
MeanSq7 = np.square(np.subtract(np.polyval(m7,x),y).mean())
print("Mean Square Error 7:",MeanSq7)
m8=np.polyfit(flatx,flaty,8) #8
MeanSq8 = np.square(np.subtract(np.polyval(m8,x),y).mean())
print("Mean Square Error 8:",MeanSq8)
axX=[1,2,3,4,5,6,7,8]
axY=[MeanSq1,MeanSq2,MeanSq3,MeanSq4,MeanSq5,MeanSq6,MeanSq7,MeanSq8]
plt.title("MSE versus order m")
plt.xlabel("Order m")
plt.ylabel("mean square error")
plt.plot(axX,axY,'ro')
plt.plot(axX,axY,'r--')
```

After comparing the Mean Square Values above i guess when the order m equals to 2 it best fits the model.
